# Ridge and Lasso Regression 

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = {
    'Area': [1000, 1500, 1800, 2400, 3000, 3500, 4000, 4500],
    'Rooms': [2, 3, 3, 4, 5, 5, 6, 7],
    'Age': [10, 5, 8, 4, 2, 7, 3, 1],
    'Price': [200000, 250000, 280000, 350000, 400000, 425000, 480000, 500000]
}
df = pd.DataFrame(data)
print(df)
```

```python
X =  df[['Area' , 'Rooms' , 'Age']]
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
lin_model = LinearRegression().fit(X_train , y_train)
lin_pred = lin_model.predict(X_test)
```

```python
print("Linear Regression coefficient : " , lin_model.coef_)
print("Linear Regession Intersept : ", lin_model.intercept_)
print("MSE : ", mean_squared_error(y_test , lin_pred))
```

```python
lasso_model = Lasso(alpha =1000 ).fit(X_train, y_train)
lasso_pred = lasso_model.predict(X_test)
```

```python
print("Linear Regression coefficient : " , lasso_model.coef_)
print("Linear Regession Intersept : ", lasso_model.intercept_)
print("MSE : ", mean_squared_error(y_test , lasso_pred))
```

```python
ridge_model = Ridge(alpha=100).fit(X_train, y_train)
ridge_pred = ridge_model.predict(X_test)
```

```python
print("Ridge Regression Coefficients:", ridge_model.coef_)
print("Ridge Regression Intercept:", ridge_model.intercept_)
print("Ridge Regression MSE:", mean_squared_error(y_test, ridge_pred))
```
