```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df['species'].unique()
```

```python
df.isnull().sum()
```

```python
df = df[df["species"] != "setosa"]
```

```python
df.head()
```

```python
df["species"] = df["species"].map({'versicolor':0 , 'virginica': 1})
```

```python
df.head()
```

```python
### split dataset into independent and depepndent feature
X = df.iloc[::-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
from sklearn.model_selection import GridSearchCV
parameter = {"penalty": ['l1', 'l2', 'elasticnet'], "C" : [1,2,3,4,5,6,10,20,30,40,50], "max_iter" : [100,200]}
```

```python
classifier_regressior = GridSearchCV(classifier, param_grid = parameter, scoring = "accuracy", cv = 5 )
```

```python
classifier_regressior.fit(X_train, y_train)
```

```python
print(classifier_regressior.best_params_)
```

```python
print(classifier_regressior.best_score_)
```

```python
### Prediction
```

```python
y_pred = classifier_regressior.predict(X_test)
```

```python
## Aaccuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
print(classification_report(y_pred, y_test))
```
